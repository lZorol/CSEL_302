#Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

#Generating Data

```python
data = pd.read_csv('Lesson_4_student_grades_dataset.csv')
print(data.head())
```

#Visualizing Data

```python
plt.scatter(data['Previous_Exam_Scores'], data['Grades'])
plt.xlabel('Independent Variable')
plt.ylabel('Dependent Variable')
plt.title('Scatter Plot of Data')
plt.show()
```

#Preparing Data for Modeling

```python
x =  data[['Previous_Exam_Scores']]
y = data[['Grades']]
```

#Fitting the Linear Regression

```python
model = LinearRegression()
model.fit(x,y)
```

#Getting Model Coefficients

```python
slope = model.coef_[0]
intercept = model.intercept_
print('Slope', slope)
print('Intercept', intercept)
```

#Visualizing the Linear Regression Line

```python
plt.scatter(x,y)
plt.plot(x, model.predict(x), color='red')
plt.xlabel('Independent Variable')
plt.ylabel('Dependent Variable')
plt.show()
```

#Making Predictions

```python
new_x = [[np.random.randint(99, 100)]]
predicted_y = model.predict(new_x)

print('Value of x:', new_x)
print('Predicted Value: ',predicted_y)
```
